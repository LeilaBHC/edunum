# Sous graphiques

Un graphique peut être découpé en multiples sous-graphics (subplots).

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Subplots
La fonction `subplot(n, m)` crée $n \times m$ sous-graphes.

```python
plt.subplots();
```

```python
plt.subplots(2, 1);
```

```python
plt.subplots(2, 1);
```

## Sélectionner un sous-graph
La fonction `subplot(n, m, i)` sélectionne le i-ième sous-graphique d'une matrice de $n \times m$ sous-graphes. 

Si le nombre des graphes est inférieur à 10 on peux les désigner directement avec `subplot(nmi)`

```python
x = [0, 1, 2]
y = [1, 3, 2]
```

```python
plt.subplot(121)
plt.plot(y, 'ro-')
plt.subplot(122)
plt.bar(x, y);
```

## Graphes superposés

```python
t = np.arange(0.0, 5.0, 0.1)

plt.subplot(211)
plt.plot(t, np.exp(-t) * np.cos(2*np.pi*t), 'bo-')

plt.subplot(212)
plt.plot(t, np.cos(2*np.pi*t), 'r--')
```
