# Graphes

Le module `matplotlib` permet d'afficher des graphes. Pour utiliser ses fonctions nous devons importer ces modules.

```python
import matplotlib.pyplot as plt
import numpy as np
```

Créons un vecteur avec des données que nous pouvons afficher.

```python
y = [2, 4, 3, 6, 3]
```

Un simple plot est obtenu avec ceci.

```python
plt.plot(y)
```

L'objet retourné est de type `Line2D`.
Dorénavant nous terminons la dernière ligne avec un point-virgule `;` pour supprimer cet output.

**Attention** pour afficher un graphe, vous devez ajouter `plt.show()` dans votre programme Python. Dans ce document-ci cette commande est inclut de façon implicit.

## Titres et légendes
Nous pouvons ajouter un titre pour le graphe ainsque que pour les deux axes x et y.

```python
plt.title('Evolution')
plt.xlabel('temps')
plt.ylabel('montant')
plt.plot(y);
```

```python
plt.plot(y, 'or', y);
```

## Coordonnées dans x

Si nous affichons un graphe à partir d'un seul vecteur, les valeurs x sont sont des entiers commençant à 0. Nous pouvons définir notre propre plage x, par exemple avec un vecteur **x** allant de -2 à +2.

Le style de la ligne est indiqué avec un texte:

- `'g^:'` green, triangle, pointillé
- `'or'` discs, red



```python
x = [-2, -1, 0, 1, 2]
plt.plot(x, y,'g^:');
```

## Propriétés de ligne

Avec des arguments à options nous pouvons changer les propriétés d'une ligne.

```python
plt.plot(x, y, linewidth=2, color='r', linestyle='--', marker='o');
```

## Histogramme

Un diagramme à barres est aussi appelé histogramme. Il représente des des informations sous forme de barres.

```python
plt.bar(x, y)
```

## Multiples lignes

Nous pouvons placer multiples lignes sur un plot.

```python
t = np.arange(0., 2., 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs:', t, t**3, 'g^:');
```

## Courbes trigonométriques

```python
t = np.linspace(0, 4*np.pi, 100)
plt.plot(t, np.sin(t), label='sin(t)')
plt.plot(t, np.cos(t), label='cos(t)')
plt.legend()
plt.grid()
```

```python
t = np.linspace(0, 4*np.pi, 100)
plt.plot(t, np.sin(t), label='sin(t)')
plt.plot(t, np.sin(t)**2, label='sin(t)**2')
plt.legend()
plt.grid()
```

## Figures de Lissajous

Les figures des Lissajous sont des combinaisons de courbes sinus et cosinus.

```python
t = np.linspace(0, 2*np.pi, 100)
plt.grid()
plt.plot(t, np.sin(3*t), 'r');
```

```python
plt.grid()
plt.plot(np.sin(t), np.cos(t), 'r');
```

```python
plt.grid()
plt.plot(np.sin(t), np.cos(3*t), 'r');
```

```python
plt.grid()
plt.plot(np.sin(2*t), np.cos(3*t), 'r');
```

## Echelle logarithmiques

```python
y = np.arange(0, 100, 2)
plt.grid()
plt.plot(y, 'o-')
plt.yscale('log')
```

```python
plt.grid()
plt.plot(y, 'o-')
plt.xscale('log')
```

```python
plt.grid()
plt.plot(y, 'o')
plt.xscale('log')
plt.yscale('log')
```
