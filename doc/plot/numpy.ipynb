# NumPy

Le module *numpy* permet le calcul numérique.

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Créer une plage

La fonction `np.arange` crée une plage

```python
np.arange(10)
```

Une plage entre deux bornes.

```python
np.arange(5, 10)
```

Avec un incrément

```python
np.arange(10, 20, 2)
```

```python
y = np.arange(10)
plt.grid()
plt.plot(y, 'o');
```

## Visualisation

```python
y = np.zeros(5)
plt.plot(range(5), y, 'o-', label='range(5)')
plt.plot(range(2, 7), y+1, 'o-', label='range(2, 7)')
plt.plot(range(0, 10, 2), y+2, 'o-', label=range(0, 10, 2))
plt.legend()
plt.ylim(-0.5, 5)
```

## Distribution uniforme

```python
y = np.random.rand(200)
plt.plot(y, 'o');
```

```python
plt.hist(y, 20);
```

## Distribution normale

```python
y = np.random.randn(200)
plt.plot(y, 'o');
```

```python
plt.hist(y, 20);
```

## Opérations

```python
plt.plot(y, 'g', 0.5*y+1, 'r');
```
