# Graphes

Le module `matplotlib` permet d'afficher des graphes. Au début du notebook nous importons ces modules.

```python
import matplotlib.pyplot as plt
import numpy as np
```

Créons un vecteur avec des données que nous pouvons afficher.

```python
y = [2, 4, 3, 6, 4]
```

Un simple plot est obtenu avec ceci.

```python
plt.plot(y)
```

L'objet retourné est de type `[<matplotlib.lines.Line2D at 0x7fac9ecef5d0>]`.
Dorénavant nous terminons la dernière ligne avec un point-virgule `;` pour supprimer cette ligne.

**Attention** pour afficher un graphe, vous devez ajouter `plt.show()` dans votre programme Python. Dans ce document-ci cette commande est inclut de façon implicit.

## Titres et légendes
Nous pouvons ajouter des légendes pour les deux axes ainsi qu'un titre

```python
plt.xlabel('temps')
plt.ylabel('montant')
plt.title('evolution')
plt.plot(y, 'or', y);
```

```python
plt.plot(y, 'or', y);
```

## Coordonnées dans x

Si nous affichons un graphe à partir d'un seul vecteur, les valeurs x sont sont des entiers commençant à 0. Nous pouvons définir notre propre plage x, par exemple avec un vecteur **x** allant de -2 à +2.

Le style de la ligne est indiqué avec un texte:

- `'g^:'` green, triangle, pointillé
- `'or'` discs, red



```python
x = [-2, -1, 0, 1, 2]
plt.plot(x, y,'g^:');
```

## Propriétés de ligne

Avec des arguments à options nous pouvons changer les propriétés d'une ligne.

```python
plt.plot(x, y, linewidth=2, color='r', linestyle='--', marker='o');
```

## Histogramme

Un diagramme à barres est aussi appelé histogramme. Il représente des des informations sous forme de barres.

```python
plt.bar(x, y);
```

## Sous-graphes

Multiples graphes peuvent être combinés. La commande `subplot(xyz)` designe le z-ième graphe de dimensions $x \times y$ 

```python
plt.subplot(121)
plt.bar(x, y)
plt.subplot(122)
plt.plot(y, 'or-');
```

Nous pouvons aussi sous-diviser en $2 \times 2$ sous-graphes.

```python
plt.subplot(221)
plt.bar(x, y)
plt.subplot(222)
plt.plot(y, 'ro-');
plt.subplot(223)
plt.plot(x, y)
plt.subplot(224)
plt.plot(y, 'og');
```

## Multiples lignes

Nous pouvons placer multiples lignes sur un plot.

```python
t = np.arange(0., 2., 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs:', t, t**3, 'g^:');
```

## Courbes sinusoidale

```python
t = np.arange(0, 2, 0.01)
s = 1 + np.sin(2 * np.pi * t)
plt.title('signal sinusoidale')
plt.ylabel('voltage (mV)')
plt.xlabel('temps (s)')
plt.grid()
plt.plot(t, s);
```

## Fréquence de lettres

Un dictionnaire est une structure qui permet de compter facilement les lettres dans une phrase.

```python
s = 'Introduction to python programming'
d = {}
for c in s:
    d[c] = d.get(c, 0) + 1
```

Ceci nous permet de dessiner un histogramme.

```python
plt.bar(d.keys(), d.values());
```

## Histogramme superposé

```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
men_std = [2, 3, 4, 1, 2]
women_std = [3, 5, 2, 3, 3]
width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, men_means, width, yerr=men_std, label='Men')
ax.bar(labels, women_means, width, yerr=women_std,
        bottom=men_means,
        label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()

plt.show()
```

## Figures de Lissajous

Les figures des Lissajous sont des combinaisons de courbes sinus et cosinus.

```python
t = np.arange(0, 10, 0.1)
plt.grid()
plt.plot(t, np.sin(3*t), 'r');
```

```python
plt.grid()
plt.plot(np.sin(t), np.cos(t), 'r');
```

```python
plt.grid()
plt.plot(np.sin(t), np.cos(3*t), 'r');
```
